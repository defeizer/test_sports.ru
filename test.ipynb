Среднее, макс и мин для всех типов:

```python
df.describe()
#  50% - медианное значение
# минимальное(min) и максимальные(max) значения и среднее(mean)
```

То же самое, но отдельно для каждого типа объектов:


```python
a = [n for n in set(df.Type)]
for i in range(len(a)):
    df1 = df[df['Type'] == a[i]]
    df1.describe().to_csv(a[i])
```

Мода для всех типов данных:       (наиболее часто встречающееся значение

```python
df.mode()
# у большинства критериев в таблице моды один показатель, но у некоторых больше(до 31). Это говорит о том,
# что такие значения встречаются в этой колонке одинаковое число раз
```
